```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import altair as alt

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
file_path = '/kaggle/input/historicbattingdata-600abs/batting_data.csv'
df = pd.read_csv(file_path)
df['AVG'] = df['AVG'].round(3)
df['OPS'] = df['OPS'].round(3)

multi = alt.selection_multi(on='mouseover', nearest=True)
title = alt.TitleParams('OPS by Year (Among Players with > 600 ABs)', anchor = 'middle')

scatter = alt.Chart(df, title = title).mark_point().encode(
    x = alt.X('yearID:Q', scale = alt.Scale(type = 'linear', domain=[1953,2024]), title = 'Year'),
    y = 'OPS:Q',
    tooltip=['full:N','AVG:Q','OPS:Q', 'yearID:Q'],
    color=alt.Color('AVG:Q').bin(maxbins=4)
).add_selection(
    multi
)
```
